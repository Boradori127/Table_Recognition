```python
def plot_counter(counter):
    items = list(counter.items())
    labels, values = zip(*items)
    
    plt.figure(figsize=(50, 6))
    plt.bar(labels, values, color='skyblue')
    plt.xlabel('Items')
    plt.ylabel('Count')
    plt.title('Count of Items in Counter')
    plt.xticks(rotation=45)
    plt.show()

def load_excel_with_openpyxl(path, sheet_name):
    total = []
    wb = openpyxl.load_workbook(path, data_only=True)
    sheet = wb[sheet_name]
    data = list(sheet.values)
    for row in data:
        total.append(row)
    return total
```

```python
def data_load(data_base_Path, data_list):
    df = []
    for idx in range(len(data_list)):
        path_name = data_base_Path + '/' + data_list[idx]
        df.append(load_excel_with_openpyxl(path_name, 'Sheet1'))
    return df
```

```python
df = data_load(data_base_Path=data_base_Path, data_list=data_path)
```
